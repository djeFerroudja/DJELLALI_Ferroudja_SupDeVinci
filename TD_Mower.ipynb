#### Traiter la colonne price afin d'avoir 3 cat√©gories : 
    - 0-300
    -301-500
    - 501- ++


```python
Category1=[]
Category2=[]
Category3=[]
for i in data_mower["price"]:
    if i >0 and i<=300 :
        Category1.append(i) 
    if i>=301 and i<=500:
        Category2.append(i) 
    if i >=501 :
        Category3.append(i) 
print("Categorie 1: ",Category1[1: 10])
print()
print("Categorie 2: ",Category2[1: 10])
print()
print("Categorie 3: ",Category3[1: 10])
```

#### Transformer la colonne product_type en colonne num√©rique avec 3 cat√©gories :

```python
#Cr√©er un disctionnaire qui a comme cl√©s: les valeurs distinctes de la colonne product_type, et les valeurs: des valeurs num√©riques.
liste1=[]
liste2 = [1,2,3]
for i in data_mower["product_type"].unique():
    liste1.append(i)
dict_prod = dict(zip(liste1, liste2))
dict_prod
```

```python
#Remplacer chaque cl√© par sa valeur: ex electrique par 2 ...
for cle, valeur in dict_prod.items():
    for j in data_mower["product_type"]:
        if cle == j:
            data_mower["product_type"]=data_mower["product_type"].replace(j, valeur)
data_mower["product_type"]
```

####  Puis cr√©er 3 colonnes (auto-port√©e, essence, √©lectrique) repr√©sentant une variable binaire avec 1 si le produit est du type de la colonne ou 0 sinon.

```python
#Methode 1 pour la colonne auto-port√©e
liste=[]
for i in data_mower["product_type"]:
    if  i == 1:
        liste.append(1)
        
    else:
        liste.append(0)
        
data_mower["auto-port√©e"]=liste
data_mower["auto-port√©e"]
```

```python
#Methode 2 Colonne √©lectrique
data_mower["√©lectrique"]=data_mower["product_type"]==2
data_mower["√©lectrique"]=data_mower["√©lectrique"].replace(False, 0)
data_mower["√©lectrique"]=data_mower["√©lectrique"].replace(True, 1)
data_mower["√©lectrique"].iloc[252:262]
```

```python
#Colonne essence
data_mower["essence"]=data_mower["product_type"]== 3
data_mower["essence"]=data_mower["essence"].replace(False, 0)
data_mower["essence"]=data_mower["essence"].replace(True, 1)
data_mower["essence"]
```
