#### 12. Transformer la colonne product_type en colonne num√©rique avec 3 cat√©gories :

```python
#Cr√©er un disctionnaire qui a comme cl√©s: les valeurs distinctes de la colonne product_type, et les valeurs: des valeurs num√©riques.
liste1=[]
liste2 = [1,2,3]
for i in data_mower["product_type"].unique():
    liste1.append(i)
dict_prod = dict(zip(liste1, liste2))
dict_prod
```

```python
#Remplacer chaque cl√© par sa valeur: ex electrique par 2 ...
for cle, valeur in dict_prod.items():
    for j in data_mower["product_type"]:
        if cle == j:
            data_mower["product_type"]=data_mower["product_type"].replace(j, valeur)
data_mower["product_type"]
```

#### 16.  Commencer √† packager son code et incorporer son script dans des fonctions pouvant √™tre r√©utilis√©es pour un autre set de donn√©es de m√™me structure (g√©n√©ralisation maximale)  

```python
# Cr√©er une nouvelle fonction qui d√©termine si la valeur en param√®tre est manquante:
def num_missing(x):
    return sum(x.isnull())

# On applique cette fonction pour chaque colonne:
print("Valeurs manquantes par colonne:")
print(data_mower.apply(num_missing, axis=0)) #axis=0 appliquer la fonction sur chaque colonne

# Puis application pour chaque ligne:
print("\nValeurs manquantes par ligne:")
print(data_mower.apply(num_missing, axis=1).head(10)) #axis=1 appliquer la fonction sur chaque ligne
```

```python
# Cr√©er une nouvelle fonction qui renvoie les valeurs uniques par colonne:
def unique_value(x):
    return x.unique()

# On applique cette fonction pour chaque colonne:
print("Valeurs uniques pour colonne:")
print(data_mower.apply(unique_value, axis=0)["quality"]) #axis=0 appliquer la fonction sur chaque colonne, affiche r√©sultat de la colonne quality
```

```python
#Cr√©er une nouvelle fonction qui remplace une valeur par une autre:
#Cette fonction nous permet de remplacer les valeurs manquantes par d'autres valeurs(moyenne par exemple),
#remplacer les valeurs caract√®res par des valeurs num√©riques ...

def replace_char(x, y, z):
    return x.replace(y,z)

# On applique cette fonction pour chaque colonne:
print("Valeurs uniques pour colonne:")
print(replace_char(data_mower["quality"],1, 'a')) #axis=0 appliquer la fonction sur chaque colonne, affiche r√©sultat de la colonne quality

replace_char(data_mower["quality"],'a', 1)
data_mower["quality"]
```
